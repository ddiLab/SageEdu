<h1 style="text-align: center;">Data Plots</h1>

This notebook simply graphs the environmental sensor data of a 0.5, 2, and 4 hour period. These .csv files can be found in `collected_data/`, labeled `tempLog_0.5hrs.csv`, `tempLog_2hrs.csv`, and `tempLog_4hrs.csv` respectively. And they follow timestamp protocol. 

---

## Setup

```python
import time
import datetime
import numpy as np
from matplotlib import pyplot as plt
```

```python
half_hour = np.loadtxt("collected_data/tempLog_0.5hrs.csv", delimiter=",")
two_hour = np.loadtxt("collected_data/tempLog_2hrs.csv", delimiter=",")
four_hour = np.loadtxt("collected_data/tempLog_4hrs.csv", delimiter=",")
```

## Helper functions

```python
"""
This helper function transforms the nanosecond format of the .csv to readable timestamps
It's used in the printData and plotData functions
"""
def toDate(t):
    return datetime.datetime.fromtimestamp(t / 1e9)
vToDate = np.vectorize(toDate)
```

## Data of half an hour

```python
half_times = vToDate(half_hour[1:, 0])
```

```python
plt.plot(half_times, half_hour[1:,1])
plt.xlabel('Time')
plt.ylabel('Temperature (C)')
plt.title("Time vs Temperature")
plt.gcf().autofmt_xdate()
```

```python
plt.plot(half_times, half_hour[1:,2])
plt.xlabel('Time')
plt.ylabel('Air Pressure (hPa)')
plt.title("Time vs Air Pressure")
plt.gcf().autofmt_xdate()
```

```python
plt.plot(half_times, half_hour[1:,3])
plt.xlabel('Time')
plt.ylabel('Humidity')
plt.title("Time vs Humidity")
plt.gcf().autofmt_xdate()
```

```python
plt.plot(half_times, half_hour[1:,4])
plt.xlabel('Time')
plt.ylabel('Gas resistance (Ohm)')
plt.title("Time vs Gas Resistance")
plt.gcf().autofmt_xdate()
```
