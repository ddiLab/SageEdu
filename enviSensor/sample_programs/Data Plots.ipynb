```python
import bme680
import time
import numpy as np
from matplotlib import pyplot as plt

sensor = bme680.BME680()

sensor.set_humidity_oversample(bme680.OS_2X)
sensor.set_pressure_oversample(bme680.OS_4X)
sensor.set_temperature_oversample(bme680.OS_8X)
sensor.set_filter(bme680.FILTER_SIZE_3)

sensor.set_gas_status(bme680.ENABLE_GAS_MEAS)
sensor.set_gas_heater_temperature(320)
sensor.set_gas_heater_duration(150)
sensor.select_gas_heater_profile(0)
```

```python
cam_data = np.loadtxt("camRunData.csv", delimiter=",")
cam2_data = np.loadtxt("camData2.csv", delimiter=",")
nano_data = np.loadtxt("nanoData.csv", delimiter=",")
```

<h2>Camera Running for an hour</h2>

```python
plt.plot(cam_data[1:, 0], cam_data[1:,1])
plt.xlabel('Minute')
plt.ylabel('Temperature (C)')
plt.title("Temperature")
```

```python
plt.plot(cam_data[1:, 0], cam_data[1:,2])
plt.xlabel('Minute')
plt.ylabel('Air Pressure (hPa)')
plt.title("Air Pressure")
```

```python
plt.plot(cam_data[1:, 0], cam_data[1:,3])
plt.xlabel('Minute')
plt.ylabel('Humidity')
plt.title("Humidity")
```

```python
plt.plot(cam_data[1:, 0], cam_data[1:,4])
plt.xlabel('Minute')
plt.ylabel('Gas resistance (Ohm)')
plt.title("Gas Resistance")
```
