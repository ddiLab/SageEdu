# Burn-In Test Results

## Summary

Burning in the sensor for a little as 10 minutes before taking readings from it seems to allow for more consistent measurements. After the burn-in process, a few minutes, about 5, should be dedicated to allow the sensor to cool down as taking readings immediately after can lead to extreme values being read in measurments such as the gas resistance, temperatue, and humidity. But after the burn-in time, the environmental sensor should be stabilized for future use. 

```python
import numpy as np
from matplotlib import pyplot as plt
```

Each of the datasets below contain the data collected from a different burn-in test.
The tests consisted of two parts: the burn-in and the data collection time.
The burn-in time consisted of taking sensor readings every second for specified number of minutes.
The burn-in times tested were 0 minutes, 10 minutes, 20 minutes, 30 minutes, 40 minutes, 50 minutes, and 60 minutes.
After the burn-in, readings were taken from the sensor once a minute for 30 minutes.
The readings taken during these 30 minutes were saved to csv files named `burn-in-test-X`, where `X` is replced by the number of minutes of the burn-in time.


```python
burn_0 = np.loadtxt("./collected_data/burn-in-test-0.csv", delimiter=",")
burn_10 = np.loadtxt("./collected_data/burn-in-test-10.csv", delimiter=",")
burn_20 = np.loadtxt("./collected_data/burn-in-test-20.csv", delimiter=",")
burn_30 = np.loadtxt("./collected_data/burn-in-test-30.csv", delimiter=",")
burn_40 = np.loadtxt("./collected_data/burn-in-test-40.csv", delimiter=",")
burn_50 = np.loadtxt("./collected_data/burn-in-test-50.csv", delimiter=",")
burn_60 = np.loadtxt("./collected_data/burn-in-test-60.csv", delimiter=",")
```

## Gas Resistance

The following are plots of the gas resistance data collected during all of the tests.

```python
plt.plot(burn_0[1:, 0], burn_0[1:,4], label='0 min')
plt.plot(burn_10[1:, 0], burn_10[1:,4], label='10 min')
plt.plot(burn_20[1:, 0], burn_20[1:,4], label='20 min')
plt.plot(burn_30[1:, 0], burn_30[1:,4], label='30 min')
plt.plot(burn_40[1:, 0], burn_40[1:,4], label='40 min')
plt.plot(burn_50[1:, 0], burn_50[1:,4], label='50 min')
plt.plot(burn_60[1:, 0], burn_60[1:,4], label='60 min')
plt.legend()

plt.xlabel('Minute')
plt.ylabel('Air resistance (Ohm)')
plt.title("Air Resistance")
```

As can be seen above, all the gas resistance readings, with the exception of the 0 min test, seemed to follow a very similar pattern.
All the readings start off high, then within about four minutes, all the readings drop down to just below 20000 Ohms.
We can see that the 0 min test did not follow this pattern, as it started low, then went up above the rest of the other tests.

```python
plt.plot(burn_0[:5, 0], burn_0[:5,4], label='0 min')
plt.plot(burn_10[:5, 0], burn_10[:5,4], label='10 min')
plt.plot(burn_20[:5, 0], burn_20[:5,4], label='20 min')
plt.plot(burn_30[:5, 0], burn_30[:5,4], label='30 min')
plt.plot(burn_40[:5, 0], burn_40[:5,4], label='40 min')
plt.plot(burn_50[:5, 0], burn_50[:5,4], label='50 min')
plt.plot(burn_60[:5, 0], burn_60[:5,4], label='60 min')

plt.legend()
plt.xlabel('Minute')
plt.ylabel('Air resistance (Ohm)')
plt.title("Air Resistance")
```

Now let's take a closer look at the first five minutes of the data. We can see that the lower the burn time, the higher the gas resistance values are.
Something interesting to note it that having at least 10 minutes of burn-in causes gas resistance values to be very high, but once the readings are taken further apart, the gas resistance readings seem to all approach the same value.
Taking a gas resistance reading requires that the sensor heats up a hot plate. The heat from this hotplate may be what's affecting the properties of the air, causing there to be higher gas resistance readings. This suggests that after burning in the sensor, it should be allowed to 'cool down' so the gas resistance stabilizes before taking actual measurments from it.

```python
plt.plot(burn_0[5:, 0], burn_0[5:,4], label='0 min')
plt.plot(burn_10[5:, 0], burn_10[5:,4], label='10 min')
plt.plot(burn_20[5:, 0], burn_20[5:,4], label='20 min')
plt.plot(burn_30[5:, 0], burn_30[5:,4], label='30 min')
plt.plot(burn_40[5:, 0], burn_40[5:,4], label='40 min')
plt.plot(burn_50[5:, 0], burn_50[5:,4], label='50 min')
plt.plot(burn_60[5:, 0], burn_60[5:,4], label='60 min')

plt.legend()
plt.xlabel('Minute')
plt.ylabel('Air resistance (Ohm)')
plt.title("Air Resistance")
```

The above plot gives us a closer look at the last 25 readings taken from the sensor. They all seem to stablize between 7000 Ohms and 14000 Ohms.

## Temperature Plots

```python
plt.plot(burn_0[1:, 0], burn_0[1:,1], label='0 min')
plt.plot(burn_10[1:, 0], burn_10[1:,1], label='10 min')
plt.plot(burn_20[1:, 0], burn_20[1:,1], label='20 min')
plt.plot(burn_30[1:, 0], burn_30[1:,1], label='30 min')
plt.plot(burn_40[1:, 0], burn_40[1:,1], label='40 min')
plt.plot(burn_50[1:, 0], burn_50[1:,1], label='50 min')
plt.plot(burn_60[1:, 0], burn_60[1:,1], label='60 min')
plt.legend()

plt.xlabel('Minute')
plt.ylabel('Temperature (C)')
plt.title("Temperature vs Time")
```

As can be seen in the plot above, the temperature values taken from the sensor follow a similar tred as the gas resistance readings. The start off high, then decrease and stabilize. The initial high temperature likely comes from the heat generated by the hotplote when the gas resistance the readings were taken during the burn-in process.

## Air Pressure Plots

```python
plt.plot(burn_0[1:, 0], burn_0[1:,2], label='0 min')
plt.plot(burn_10[1:, 0], burn_10[1:,2], label='10 min')
plt.plot(burn_20[1:, 0], burn_20[1:,2], label='20 min')
plt.plot(burn_30[1:, 0], burn_30[1:,2], label='30 min')
plt.plot(burn_40[1:, 0], burn_40[1:,2], label='40 min')
plt.plot(burn_50[1:, 0], burn_50[1:,2], label='50 min')
plt.plot(burn_60[1:, 0], burn_60[1:,2], label='60 min')
plt.legend()

plt.xlabel('Minute')
plt.ylabel('Air Pressure (hPa)')
plt.title("Air Pressure vs Time")
```

The air pressure did not significantly change over time throughout the tests, indicating that the gas resistance readings don't have an effect on the air pressure around the sensor and burn-in time doesn't matter for it.
