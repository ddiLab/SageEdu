```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
burn_0 = np.loadtxt("./burnInTests/burn-in-test-0.csv", delimiter=",")
burn_10 = np.loadtxt("./burnInTests/burn-in-test-10.csv", delimiter=",")
burn_20 = np.loadtxt("./burnInTests/burn-in-test-20.csv", delimiter=",")
burn_30 = np.loadtxt("./burnInTests/burn-in-test-30.csv", delimiter=",")
burn_40 = np.loadtxt("./burnInTests/burn-in-test-40.csv", delimiter=",")
burn_50 = np.loadtxt("./burnInTests/burn-in-test-50.csv", delimiter=",")
burn_60 = np.loadtxt("./burnInTests/burn-in-test-60.csv", delimiter=",")
```

## Gas Resistance Plots

```python
plt.plot(burn_0[1:, 0], burn_0[1:,4], label='0 min')
plt.plot(burn_10[1:, 0], burn_10[1:,4], label='10 min')
plt.plot(burn_20[1:, 0], burn_20[1:,4], label='20 min')
plt.plot(burn_30[1:, 0], burn_30[1:,4], label='30 min')
plt.plot(burn_40[1:, 0], burn_40[1:,4], label='40 min')
plt.plot(burn_50[1:, 0], burn_50[1:,4], label='50 min')
plt.plot(burn_60[1:, 0], burn_60[1:,4], label='60 min')
plt.legend()

plt.xlabel('Minute')
plt.ylabel('Air resistance (Ohm)')
plt.title("Air Resistance")
```
