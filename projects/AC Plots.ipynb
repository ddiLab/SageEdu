```python
import numpy as np
import datetime
from matplotlib import pyplot as plt
audio2 = np.loadtxt("./volumeLog_2hr.csv", delimiter=",")
temp2 = np.loadtxt("../enviSensor/tempLog_2hr.csv", delimiter=",")
audio4 = np.loadtxt("./volumeLog_4hrs.csv", delimiter=",")
temp4 = np.loadtxt("../enviSensor/tempLog_4hrs.csv", delimiter=",")
```

```python
def toDate(t):
    return datetime.datetime.fromtimestamp(t / 1e9)
vToDate = np.vectorize(toDate)
```

## 2 hours

```python
plt.plot(vToDate(audio2[1:, 0]), audio2[1:,1], label='Avg')
plt.plot(vToDate(audio2[1:, 0]), audio2[1:,2], label='Max')
plt.plot(vToDate(audio2[1:, 0]), audio2[1:,3], label='Min')

plt.legend()
plt.xlabel('Time')
plt.ylabel('Volume')
plt.title("Volume over 2hrs")
plt.gcf().autofmt_xdate()
```

```python
fig,ax = plt.subplots()
ax.plot(vToDate(audio2[1:, 0]), audio2[1:,1], color="green", label='Avg')
ax.set_xlabel("Time",fontsize=14)
ax.set_ylabel("Avg Volume",color="green",fontsize=14)

ax2=ax.twinx()
ax2.plot(vToDate(temp2[1:, 0]), temp2[1:,1], color="blue")
ax2.set_ylabel("Temperature (C)",color="blue",fontsize=14)
plt.gcf().autofmt_xdate()
plt.show()
```
