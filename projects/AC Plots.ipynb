```python
import numpy as np
from matplotlib import pyplot as plt
audioData = np.loadtxt("./volumeLog_2hr.csv", delimiter=",")
tempData = np.loadtxt("../enviSensor/tempLog_2hr.csv", delimiter=",")
```

```python
plt.plot(audioData[1:, 0], audioData[1:,1], label='Avg')
plt.plot(audioData[1:, 0], audioData[1:,2], label='Max')
plt.plot(audioData[1:, 0], audioData[1:,3], label='Min')

plt.legend()
plt.xlabel('Time')
plt.ylabel('Volume')
plt.title("Volume over 2hrs")
```

```python
plt.plot(audioData[1:, 0], audioData[1:,1], label='Avg')
plt.xlabel('Time')
plt.ylabel('Volume')
plt.title("Avg Volume over 2hrs")
```

```python
plt.plot(tempData[1:, 0], tempData[1:,1], label='Avg')
plt.xlabel('Time')
plt.ylabel('Temperature (C)')
plt.title("Temperature over 2hrs")
```
