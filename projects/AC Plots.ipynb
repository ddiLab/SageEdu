```python
import numpy as np
from matplotlib import pyplot as plt
audio2 = np.loadtxt("./volumeLog_2hr.csv", delimiter=",")
temp2 = np.loadtxt("../enviSensor/tempLog_2hr.csv", delimiter=",")
audio4 = np.loadtxt("./volumeLog_4hrs.csv", delimiter=",")
temp4 = np.loadtxt("../enviSensor/tempLog_4hrs.csv", delimiter=",")
```

## 2 hours

```python
plt.plot(audio2[1:, 0], audio2[1:,1], label='Avg')
plt.plot(audio2[1:, 0], audio2[1:,2], label='Max')
plt.plot(audio2[1:, 0], audio2[1:,3], label='Min')

plt.legend()
plt.xlabel('Time')
plt.ylabel('Volume')
plt.title("Volume over 2hrs")
```
