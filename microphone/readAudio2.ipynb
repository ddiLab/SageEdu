#### Now, import the read function to get data from the .wav file

```python
from scipy.io.wavfile import read
```

```python
# fs is the sample rate
# x is the data from the .wav file
(fs, x) = read('waves.wav')
```

#### Interpreting the data:

```python
print(f"The value of fs: {fs} \n The value of x: {x}")
```

```python
duration = len(x)/float(fs)
print(f"Duration in seconds of the file: {duration}")
```

#### We will import numpy and matplotlib t plot the data

```python
import matplotlib.pyplot as plt
import numpy as np
```

#### Method 1 of plotting the audio:

```python
time = np.linspace(0, len(x) / float(fs), num = len(x))
```

```python
plt.figure(1)
plt.title('Waves')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.plot(time, x)
```

#### Method 2 of plotting audio:

```python
time2 = np.arange(0, len(x)/fs, 1/fs)
```

```python
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Waves')
plt.plot(time2, x)
```
